```python
dic={}
subject =['Maths','English','Python']

for i in subject:
    dic[i]=int(input(f"Enter the marks for {i}: "))


sum=0
flag=True
for i in dic.values():
    if i<33:
        flag=False
    sum+=i
if sum/3 and flag:
    print("Pass")
else:
    print("Fail")

```
